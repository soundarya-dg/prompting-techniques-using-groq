### Library imports

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
```

### Model selection and tokenizer setup

```python
# Specify the model name
# Here we use "microsoft/phi-2" as an example of a reasoning-capable model
model_name = "microsoft/phi-2"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float16,
    device_map="auto"
)
```

### 1. Zero Shot Prompting

In this technique, you give the model only an instruction with no examples. The model needs to figure out the task on its own.

```python
# Define the prompt
prompt = """Translate the following English text to French: 'Hello, how are you?'
And then translate the French text back to Spanish.

Output format:
French: <French translation>
Spanish: <Spanish translation>
"""

# Tokenize the input prompt and generate the output
inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

outputs = model.generate(
    **inputs,
    max_new_tokens=120,     # Tokens are limited for demonstration; adjust as needed
    temperature=0.7         # Adjust temperature for creativity in responses
)

print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```
